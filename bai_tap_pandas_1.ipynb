```python
import pandas as pd
import numpy as np
b = pd.Series(np.random.randint(1,10,[10]))
l2 = ['a','b','c','d','e','f','g','h','l','m']
b.index = l2
print(b)
print("chi so a: ",b.loc['a'])
print("chi so c: ",b.loc['c'])
print("chi so e: ",b.loc['e'])
```
